```python
employees = [
    {"ID":101,"Name":"Asha","Department":"HR","Salary":50000},
    {"ID":102,"Name":"Rahul","Department":"IT","Salary":65000},
    {"ID":103,"Name":"Priya","Department":"Finance","Salary":55000},
    {"ID":104,"Name":"Kiran","Department":"IT","Salary":70000},
    {"ID":105,"Name":"Meena","Department":"HR","Salary":45000}
]

for emp in employees:
    print(emp)
```

```python
for emp in employees:
    if emp["Department"] == "IT":
        print(emp)
```

```python
for emp in employees:
    if emp["Salary"] > 50000:
        print(emp["Name"], emp["Salary"])
```

```python
highest = max(emp["Salary"] for emp in employees)

print("Highest Salary:", highest)
```

```python
lowest = min(emp["Salary"] for emp in employees)

print("Lowest Salary:", lowest)
```

```python
total = 0

for emp in employees:
    total += emp["Salary"]

average = total / len(employees)

print("Average Salary:", average)
```

```python
department_count = {}

for emp in employees:
    dept = emp["Department"]

    if dept in department_count:
        department_count[dept] += 1
    else:
        department_count[dept] = 1

print(department_count)
```
