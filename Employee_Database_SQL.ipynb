```python
import sqlite3
import pandas as pd

# Connect to SQLite in-memory database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# ===========================
# Create Employees Table
# ===========================
cursor.execute("""
CREATE TABLE employees (
    employee_id INTEGER PRIMARY KEY,
    employee_name TEXT,
    age INTEGER,
    gender TEXT,
    department_id INTEGER,
    salary REAL,
    experience INTEGER,
    performance_rating INTEGER,
    joining_date DATE,
    work_mode TEXT
);
""")

# ===========================
# Create Departments Table
# ===========================
cursor.execute("""
CREATE TABLE departments (
    department_id INTEGER PRIMARY KEY,
    department_name TEXT,
    manager_name TEXT,
    location TEXT
);
""")

# ===========================
# Create Projects Table
# ===========================
cursor.execute("""
CREATE TABLE projects (
    project_id INTEGER PRIMARY KEY,
    employee_id INTEGER,
    project_name TEXT,
    project_status TEXT,
    hours_worked INTEGER,
    FOREIGN KEY(employee_id) REFERENCES employees(employee_id)
);
""")

# ===========================
# Insert Departments
# ===========================
cursor.execute("""
INSERT INTO departments VALUES
(1,'HR','Kavitha','Chennai'),
(2,'IT','Suresh','Bangalore'),
(3,'Finance','Anita','Hyderabad'),
(4,'Marketing','Ramesh','Mumbai'),
(5,'Sales','Deepak','Pune');
""")

# ===========================
# Insert Employees
# ===========================
cursor.execute("""
INSERT INTO employees VALUES
(101,'Asha',24,'Female',1,45000,2,4,'2022-01-10','Hybrid'),
(102,'Rahul',28,'Male',2,70000,5,5,'2020-05-12','Remote'),
(103,'Priya',26,'Female',3,60000,4,3,'2021-08-18','Office'),
(104,'Kiran',30,'Male',2,80000,7,5,'2019-04-22','Hybrid'),
(105,'Meena',25,'Female',4,50000,3,4,'2023-02-01','Office');
""")

# ===========================
# Insert Projects
# ===========================
cursor.execute("""
INSERT INTO projects VALUES
(201,101,'Employee Engagement Dashboard','Completed',120),
(202,102,'AI Recruitment System','In Progress',180),
(203,103,'Payroll Analytics','Completed',140),
(204,104,'Talent Intelligence Dashboard','In Progress',220),
(205,105,'Marketing Performance Report','Completed',100);
""")

conn.commit()

# ===========================
# Display Tables
# ===========================
print("Employees Table")
print(pd.read_sql("SELECT * FROM employees", conn))

print("\nDepartments Table")
print(pd.read_sql("SELECT * FROM departments", conn))

print("\nProjects Table")
print(pd.read_sql("SELECT * FROM projects", conn))
```
