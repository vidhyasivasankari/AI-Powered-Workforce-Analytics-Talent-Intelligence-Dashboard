```python
import pandas as pd

employee = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Asha', 'Rahul', 'Priya', 'Kiran', 'Meena'],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR'],
    'Salary': [40000, 55000, 50000, 65000, 45000]
}

df_emp = pd.DataFrame(employee)
display(df_emp)
```

```python
department = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing'],
    'Manager': ['Anita', 'Ramesh', 'Suresh', 'Deepa']
}

df_dept = pd.DataFrame(department)
display(df_dept)
```

```python
inner_join = pd.merge(df_emp, df_dept, on='Department', how='inner')
display(inner_join)
```

```python
left_join = pd.merge(df_emp, df_dept, on='Department', how='left')
display(left_join)
```

```python
right_join = pd.merge(df_emp, df_dept, on='Department', how='right')
display(right_join)
```

```python
outer_join = pd.merge(df_emp, df_dept, on='Department', how='outer')
display(outer_join)
```

```python
import sqlite3

conn = sqlite3.connect("employee.db")
cursor = conn.cursor()

print("Database Connected Successfully")
```

```python
cursor.execute("""
SELECT * FROM Employee
WHERE Salary > 50000
""")
print(cursor.fetchall())
```

```python
cursor.execute("""
SELECT * FROM Employee
WHERE Department = 'IT'
AND Salary > 50000
""")

print(cursor.fetchall())
```

```python
cursor.execute("""
SELECT *
FROM Employee
WHERE Department = 'HR'
OR Salary > 60000
""")

print(cursor.fetchall())
```
